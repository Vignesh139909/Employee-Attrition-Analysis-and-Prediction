```python
import pandas as pd
import numpy as np
```

```python
employee_data = pd.read_csv(r"C:\Users\Vignesh\OneDrive\Desktop\Datascience\PROJECT1\Employee\Employee-Attrition.csv")
employee_data.head()
```

```python
df = employee_data.copy()
df.head()
```

```python
print(f"DataFrame shape: {df.shape}")
print(f"Number of rows: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
```

```python
df.info()
```

```python
for column in employee_data.columns:
    print(f"{column} - Number of unique values : {employee_data[column].nunique()}")
    print("=============================================================")
```

```python
df.select_dtypes(np.number).sample(5)
```

```python
df["Education"] = df["Education"].replace({1:"High School",2:"Diploma",3:"Graduate",4:"Postgraduate",5:"Doctorate"})
df["JobInvolvement"] = df["JobInvolvement"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df["JobSatisfaction"] = df["JobSatisfaction"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df["PerformanceRating"] = df["PerformanceRating"].replace({1:"Low",2:"Good",3:"Excellent",4:"Outstanding"})
df["RelationshipSatisfaction"] = df["RelationshipSatisfaction"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df["WorkLifeBalance"] = df["WorkLifeBalance"].replace({1:"Bad",2:"Good",3:"Better",4:"Best"})
df["JobLevel"] = df["JobLevel"].replace({1:"Entry_Level",2:"Junior_Level",3:"Mid_Level",4:"Senior_Level",5:"Executive_Level"})
df["EnvironmentSatisfaction"] = df["EnvironmentSatisfaction"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
```

```python
df.select_dtypes(include=['object']).sample(5)
```

```python
missing_df = df.isnull().sum().to_frame().rename(columns={0:"Total No. of Missing Values"})
missing_df
```

```python
categorical_features = []
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 30:
        categorical_features.append(column)
        print(f"{column} : {df[column].unique()}")
        print(df[column].value_counts())
        print("====================================================================================")
categorical_features.remove('Attrition')
```

```python
df.drop(['EmployeeCount', 'Over18', 'StandardHours','EmployeeNumber'], axis="columns", inplace=True)
```
